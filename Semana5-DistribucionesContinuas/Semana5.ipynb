# Distribucion Normal: Ejemplo 1

En la fabricación de cereales, se contempla el peso de las cajas como una variable aleatoria con distribución normal. Una marca reconocida presenta cajas cuyo peso promedio es de 1000 gramos y una desviación estándar de 20 gramos. Se selecciona una caja al azar con el propósito de realizar un análisis detallado.

Determinar la probabilidad de que pese menos de 1025 gramos.

Hallar la probabilidad de que pese 989 gramos o más.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 1000
desviacion_estandar = 20

# Valor para el cual queremos calcular la probabilidad (X < 1025)
valor_limite1 = 1025
valor_limite2 = 989

# Calcula la probabilidad utilizando la función de distribución acumulativa (CDF)
probabilidad1 = norm.cdf(valor_limite1, loc=media, scale=desviacion_estandar)
probabilidad2 = (1-norm.cdf(valor_limite2, loc=media, scale=desviacion_estandar))

print(f"La probabilidad de que pese  menos de {valor_limite1} onzas  es: {probabilidad1:.1%}")
print(f"La probabilidad de que pese  mas de {valor_limite2} onzas  es: {probabilidad2:.1%}")
```

```python
# Generar datos para el gráfico
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 1000)
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Grafica la distribución normal
plt.plot(x, y, label='Distribución Normal')

# Sombréa la región P(X < 1025)
x_fill = np.linspace(media - 3 * desviacion_estandar, valor_limite1, 1000)
y_fill = norm.pdf(x_fill, loc=media, scale=desviacion_estandar)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='P(X < 1025)')

# Resalta la media
plt.axvline(media, color='red', linestyle='--', label='Media')

# Configura el gráfico
plt.title('Distribución Normal y P(X < 1025)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Muestra el porcentaje asociado
plt.text(1002, 0.015, f'P(X < 1025) = {probabilidad1:.2%}', fontsize=10, color='darkblue')

# Muestra el gráfico
plt.show()
```

# Distribucion Normal: Ejemplo 2


Suponga que una máquina de café  ofrece tazas de café de 12 onzas. La cantidad real de café vertida en la taza por la máquina varía según una distribución normal, con una media igual a 13 onzas y una desviación estándar igual a 0,6 onzas. Para cada pregunta que se presenta a continuación, es necesario crear un gráfico en el cual se destaque la media, sombrear la región correspondiente a la respuesta y calcular el porcentaje asociado.


a. ¿Qué porcentaje de tazas se llenará con menos de 12 onzas?

b. ¿Qué porcentaje de tazas se llenará con más de 12.5 onzas?

c. ¿Qué porcentaje de tazas tendrá entre 12 y 13 onzas de café?

```python
from scipy.stats import norm
# Parámetros de la distribución normal
media = 13
desviacion_estandar = 0.6

# Valor para el cual queremos calcular la probabilidad (X < 12)
valor_limite1 = 12
valor_limite2 = 12.5
valor_limite3 = 13

# Calcula la probabilidad utilizando la función de distribución acumulativa (CDF)
probabilidad1 = norm.cdf(valor_limite1, loc=media, scale=desviacion_estandar)
probabilidad2 = (1-norm.cdf(valor_limite2, loc=media, scale=desviacion_estandar))
probabilidad3 = norm.cdf(13, loc=media, scale=desviacion_estandar)-norm.cdf(12, loc=media, scale=desviacion_estandar)

# Imprime el resultado
print(f"El porcentaje de tazas que se llenará con menos de {valor_limite1} onzas  es: {probabilidad1:.1%}")
print(f"El porcentaje de tazas que se llenará con mas de {valor_limite2} onzas  es: {probabilidad2:.1%}")
print(f"El porcentaje de tazas que tendra entre {valor_limite1} y {valor_limite3} es: {probabilidad3:.1%}")
```

```python
#Código para el gráfico P(X<12)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 13
desviacion_estandar = 0.6

# Valor para el cual queremos calcular la probabilidad (X < 12)
valor_limite = 12

# Calcula la probabilidad utilizando la función de distribución acumulativa (CDF)
probabilidad = norm.cdf(valor_limite, loc=media, scale=desviacion_estandar)

# Genera datos para el gráfico
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 1000)
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Grafica la distribución normal
plt.plot(x, y, label='Distribución Normal')

# Sombréa la región P(X < 12)
x_fill = np.linspace(media - 3 * desviacion_estandar, valor_limite, 1000)
y_fill = norm.pdf(x_fill, loc=media, scale=desviacion_estandar)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='P(X < 12)')

# Resalta la media
plt.axvline(media, color='red', linestyle='--', label='Media')

# Configura el gráfico
plt.title('Distribución Normal y P(X < 12)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Muestra el porcentaje asociado
plt.text(11.8, 0.3, f'P(X < 12) = {probabilidad:.2%}', fontsize=10, color='darkblue')

# Muestra el gráfico
plt.show()
```

```python
#Código para el gráfico P(12 < X < 13)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 13
desviacion_estandar = 0.6

# Valores para los límites de la región P(12 < X < 13)
limite_inferior = 12
limite_superior = 13

# Calcula las probabilidades utilizando la función de distribución acumulativa (CDF)
probabilidad_inferior = norm.cdf(limite_inferior, loc=media, scale=desviacion_estandar)
probabilidad_superior = norm.cdf(limite_superior, loc=media, scale=desviacion_estandar)

# Calcula la probabilidad de la región P(12 < X < 13)
probabilidad_region = probabilidad_superior - probabilidad_inferior

# Genera datos para el gráfico
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 1000)
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Grafica la distribución normal
plt.plot(x, y, label='Distribución Normal')

# Sombréa la región P(12 < X < 13)
x_fill = np.linspace(limite_inferior, limite_superior, 1000)
y_fill = norm.pdf(x_fill, loc=media, scale=desviacion_estandar)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='P(12 < X < 13)')

# Resalta la media
plt.axvline(media, color='red', linestyle='--', label='Media')

# Etiquetas para el gráfico
plt.title('Distribución Normal y P(12 < X < 13)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Muestra el porcentaje asociado
plt.text(12.2, 0.3, f'P(12 < X < 13) = {probabilidad_region:.1%}', fontsize=10, color='darkblue')

# Muestra el gráfico
plt.show()
```
