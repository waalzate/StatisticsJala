# Distribución Normal: Ejemplo 

Suponga que una máquina de café sirva tazas de café de 12 gramos. La cantidad real de café colocada en la taza por la máquina varía según una distribución normal, con una media igual a 13 gramos y una desviación estándar igual a 0,6 gramos. Para cada cuestión presentada a continuación, es necesario crear un gráfico destacando la media, sombreando la región correspondiente a la respuesta y calculando el porcentaje asociado. ​

a. ¿Qué porcentaje de tazas se llenará con menos de 12 gramos? ​

b. ¿Qué porcentaje de tazas se llenará con más de 12,5 gramos? ​

c. ¿Qué porcentaje de tazas tendrá entre 12 y 13 gramos de café?

```python
from scipy.stats import norm

# Parámetros de la distribución normal
media = 13
desviacion_estandar = 0.6
limite_1 = 12

prob_menor_12=norm.cdf(limite_1,loc=media, scale = desviacion_estandar)
print(f"El porcentaje de tazas con menos de {limite_1} gramos es: {prob_menor_12:.3f}")
```

```python
prob_menor_12
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 13
desviacion_estandar = 0.6

# Valor para el que queremos calcular la probabilidad (X<12)
valor_limite = 12

# Calcula la probabilidad utilizando la función de distribución acumulada (CDF)
probabilidad = norm.cdf(valor_limite, loc=media, scale=desviacion_estandar)

# Genera los datos para el gráfico
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 1000)
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Gráfico de la distribución normal
plt.plot(x, y, label='Distribución Normal')

# Sombra de la región P(X < 12)
x_fill = np.linspace(media - 3 * desviacion_estandar, valor_limite, 1000)
y_fill = norm.pdf(x_fill, loc=media, scale=desviacion_estandar)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='P(X < 12)')

# Resaltar la media
plt.axvline(media, color='red', linestyle='--', label='Media')

# Configurar el gráfico
plt.title('Distribución Normal y P(X < 12)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Mostrar el porcentaje
plt.text(11.8, 0.3, f'P(X < 12) = {probabilidad:.2%}', fontsize=10, color='darkblue')

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
media = 13
desviacion_estandar = 0.6

# Valores para los límites de la región P(12 < X < 13)
limite_inferior = 12
limite_superior = 13

# Calcula las probabilidades utilizando la función de distribución acumulada (CDF)
probabilidad_inferior = norm.cdf(limite_inferior, loc=media, scale=desviacion_estandar)
probabilidad_superior = norm.cdf(limite_superior, loc=media, scale=desviacion_estandar)

# Calcula la probabilidad de la región P(12 < X < 13)
probabilidad_region = probabilidad_superior - probabilidad_inferior

# Genera datos para el gráfico
x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 1000)
y = norm.pdf(x, loc=media, scale=desviacion_estandar)

# Gráfico de la distribución normal
plt.plot(x, y, label='Distribución Normal')

# Sombra de la región P(12 < X < 13)
x_fill = np.linspace(limite_inferior, limite_superior, 1000)
y_fill = norm.pdf(x_fill, loc=media, scale=desviacion_estandar)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='P(12 < X < 13)')

# Resaltar la media
plt.axvline(media, color='red', linestyle='--', label='Media')

# Configurar el gráfico
plt.title('Distribución Normal y P(12 < X < 13)')
plt.xlabel('X')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Mostrar el porcentaje
plt.text(12.2, 0.3, f'P(12 < X < 13) = {probabilidad_region:.2%}', fontsize=10, color='darkblue')

# Mostrar el gráfico
plt.show()
```

## Ejemplo práctico ( Distribución normal)

Usaremos la siguiente basede datos de Kaggle

https://www.kaggle.com/datasets/jasmeet0516/air-quality


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Cargamos la base de datos previamente guardada en la base de datos:
df = pd.read_csv('air_quality.csv')

df
```

```python
# Filtrar los datos para la medida específica (PM2.5)
datos_pm25 = df[df['Name'] == 'Fine particles (PM 2.5)']
```

```python
# Consideramos los valores de interes
valores_pm25= datos_pm25['Data Value']
```

```python
valores_pm25.describe()
```

```python
# Graficar la distribución de la concentración de PM2.5
plt.figure(figsize=(8, 5))
plt.hist(valores_pm25, bins=30, color='blue', alpha=0.6)
plt.title('Distribución de la Concentración de PM2.5')
plt.xlabel('Concentración de PM2.5 (µg/m³)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
#Realmente no observamos la campana de gaus, pero podemos aproximarla por medio de una 
# distribución normal.

# Calcular estadísticas básicas
media_pm25 = np.mean(valores_pm25)
desvio_estandar_pm25 = np.std(valores_pm25)

print(f'Media de la concentración de PM2.5: {media_pm25:.2f} µg/m³')
print(f'Desviación estándar de la concentración de PM2.5: {desvio_estandar_pm25:.2f} µg/m³')
```

```python
#supongamos que los datos siguen una distribución normal:
# Graficar la distribución de las concentraciones simuladas y la función de densidad
plt.figure(figsize=(8, 5))
plt.hist(valores_pm25, bins=30, color='blue', alpha=0.6, density=True, label='Datos reales')

# Calcular la función de densidad de la distribución normal con media 9.17 y desviación estándar 2.10
x = np.linspace(0, 30, 1000)
y = norm.pdf(x, loc=9.17, scale=2.10)
plt.plot(x, y, color='red', label='Densidad Normal aproximada')

plt.title('Distribución  de la Concentración de PM2.5')
plt.xlabel('Concentración de PM2.5 (µg/m³)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calcular la probabilidad de que la concentración de PM2.5 exceda los 12 µg/m³


limite_pm25 = 12
probabilidad_exceder_limite = 1 - norm.cdf(limite_pm25, media_pm25, desvio_estandar_pm25)

print(f'Probabilidad de que la concentración de PM2.5 exceda los {limite_pm25} µg/m³: {probabilidad_exceder_limite:.3%}')
```
